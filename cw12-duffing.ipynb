# Summary of the problem

Consider a ball rolling in a double-well potential described by $V(x)=\frac{x^4}{4}-\frac{x^2}{2}$.  The potential makes a plot shaped like a sombrero.  The plot is shown below.

```python
def potential(x):
    return x**4/4-x**2/2
t = np.linspace(-1.7,1.7,100)
plt.plot(t,potential(t))
plt.xlabel("position")
plt.ylabel("potential")
plt.title("'Sombrero' potential")
plt.show()
```

Assusme the ball has mass $m$, the drag coefficient is $\nu$, and the hat is being shaken at frequency $\omega$ with constant force amplitude $F$.  Then, this non-linear oscillator can be desribed by the following coupled differential equations.
$$\dot{x}(t) = y(t)$$
$$m\dot{y}(t) = -\nu y(t)+x(t)-x^3(t)+F\cos (\omega t)$$

For the purpose of this problem, we set $m=1$, $\omega = 1$, and $\nu = 0.25$.  Then, our equations become:
$$\dot{x}(t) = y(t)$$
$$\dot{y}(t) = -0.25 y(t)+x(t)-x^3(t)+F\cos ( t)$$

We will explore how this system evolves over time using numerical methods.

# Problem 1
We investigate the case where $F=0.18, \hspace{2mm}x(0)=-0.9, \hspace{2mm}y(0)=0$.

```python
t,x,y = sombrero.rk(a=0, b=50, x0=-0.9, y0=0, nu=0.25, F=0.18)
```

```python
sombrero.plot(x,t,'Motion of the oscillator')
```

```python
sombrero.parametric_plot(x,y,'Phase space of this duffing oscillator')
```

```python
sombrero.scatter_plot(x,y,50,"Poincare section of this duffing oscillator")
```

For this duffing oscillator, equilibrium seems to be reached relatively quickly in the left potential well.  This can be seen most clearly in the $t$ vs. $x$ plot of the system.  One can see from the phase space diagram and the Poincare section that some sort of resonance is achieved.  In the phase space diagram, a certain egg-shaped pattern is thickly traced, indicating that the behavior of the system was approaching some sort of equilibrium behavior near that area.  Also, on the Poincare section diagram, one can see that a very specific value of $(x,\dot{x})$ is approached as $t$ approaches $2n\pi$ for $n=0,1,2,\dots$  This point, roughly $(-0.815,0.04)$, must be some sort of attractor for the system.
